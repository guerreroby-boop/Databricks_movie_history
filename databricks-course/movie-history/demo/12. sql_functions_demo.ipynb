###Scalar Functions (Funciones Escalares)

```sql
USE movie_silver
```

```sql
SELECT *, CONCAT(title, ' - ', release_date) AS title_release
FROM movies;
```

```sql
SELECT *, 
      SPLIT(name, '\\s+')[0] forname, 
      SPLIT(name, '\\s+')[1] surnamne
FROM persons;
```

```sql
SELECT *, current_timestamp()
FROM movies;
```

```sql
SELECT *, date_format(release_date, 'dd-MM-yyyy')
FROM movies
```

```sql
SELECT *, date_add(release_date, 1)
FROM movies;
```

###Agregaciones

```sql
SELECT COUNT(1)
FROM movies;
```

```sql
SELECT MAX(release_date)
FROM movies;
```

```sql
SELECT year_release_date, COUNT(1), SUM(budget)
FROM movies
GROUP BY year_release_date
HAVING COUNT(1) > 200;
```

###Window Functions

```sql
SELECT title, year_release_date, release_date, 
  RANK() OVER(PARTITION BY year_release_date ORDER BY release_date) AS rank
FROM movies
WHERE year_release_date IS NOT NULL
ORDER BY release_date;
```
