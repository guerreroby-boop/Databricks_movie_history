###Read & Write en Delta Lake
1. Escribir datos en Delta Lake (Managed Table)
2. Escribi datos en Delta Lake (External Table)
3. Leer datos desde Delta Lake (Table)
4. Leer datos desde Delta Lake (File)

```python
%sql
CREATE SCHEMA IF NOT EXISTS movie_demo
MANAGED LOCATION "abfss://demo@moviehistorybymg.dfs.core.windows.net"
```

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType, DateType
```

```python
movie_schema = StructType([
    StructField("movieId", IntegerType(), False),
    StructField("title", StringType(), True),
    StructField("budget", DoubleType(), True),
    StructField("homePage", StringType(), True),
    StructField("overview", StringType(), True),
    StructField("popularity", DoubleType(), True),
    StructField("yearReleaseDate", IntegerType(), True),
    StructField("releaseDate", DateType(), True),
    StructField("revenue", DoubleType(), True),
    StructField("durationTime", IntegerType(), True),
    StructField("movieStatus", StringType(), True),
    StructField("tagline", StringType(), True),
    StructField("voteAverage", DoubleType(), True),
    StructField("voteCount", IntegerType(), True)
])
```

```python
movie_df = spark.read \
    .option("header", True) \
    .schema(movie_schema) \
    .csv("abfss://bronze@moviehistorybymg.dfs.core.windows.net/2024-12-30/movie.csv")
```

```python
display(movie_df)
```

#### Cargar datos externos desde formato csv a una tabla administrada

```python
movie_df.write.format("delta").mode("overwrite").saveAsTable("movie_demo.movies_managed")
```

```python
%sql
Select * from movie_demo.movies_managed
```

#### Cargar datos externos desde formato csv a un archivo externo

```python
movie_df.write.format("delta").mode("overwrite").save("abfss://demo@moviehistorybymg.dfs.core.windows.net/movies_external")
```

```python
%sql
CREATE TABLE movie_demo.movies_external
USING DELTA
LOCATION "abfss://demo@moviehistorybymg.dfs.core.windows.net/movies_external"
```

```python
%sql
select * from movie_demo.movies_external
```

#### Cargar datos externos desde formato delta a un data frame

```python
movies_external_df = spark.read.format("delta").load("abfss://demo@moviehistorybymg.dfs.core.windows.net/movies_external")
```

###Merge / Upsert en Delta Lake

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType, DateType
```

```python
movie_schema = StructType([
    StructField("movieId", IntegerType(), False),
    StructField("title", StringType(), True),
    StructField("budget", DoubleType(), True),
    StructField("homePage", StringType(), True),
    StructField("overview", StringType(), True),
    StructField("popularity", DoubleType(), True),
    StructField("yearReleaseDate", IntegerType(), True),
    StructField("releaseDate", DateType(), True),
    StructField("revenue", DoubleType(), True),
    StructField("durationTime", IntegerType(), True),
    StructField("movieStatus", StringType(), True),
    StructField("tagline", StringType(), True),
    StructField("voteAverage", DoubleType(), True),
    StructField("voteCount", IntegerType(), True)
])
```

```python
movie_day1_df = spark.read \
    .option("header", True) \
    .schema(movie_schema) \
    .csv("abfss://bronze@moviehistorybymg.dfs.core.windows.net/2024-12-30/movie.csv") \
    .filter("yearReleaseDate < 2000") \
    .select("movieId", "title", "yearReleaseDate", "releaseDate", "durationTime")
```

```python
display(movie_day1_df)
```

```python
movie_day1_df.createOrReplaceTempView("movie_day1")
```

```python
from pyspark.sql.functions import upper

movie_day2_df = spark.read \
    .option("header", True) \
    .schema(movie_schema) \
    .csv("abfss://bronze@moviehistorybymg.dfs.core.windows.net/2024-12-30/movie.csv") \
    .filter("yearReleaseDate BETWEEN 1998 AND 2005") \
    .select("movieId", upper("title").alias("title"), "yearReleaseDate", "releaseDate", "durationTime")
```

```python
movie_day2_df.createOrReplaceTempView("movie_day2")
```

```python
display(movie_day2_df)
```

```python
from pyspark.sql.functions import upper
movie_day3_df = spark.read \
    .option("header", True) \
    .schema(movie_schema) \
    .csv("abfss://bronze@moviehistorybymg.dfs.core.windows.net/2024-12-30/movie.csv") \
    .filter("yearReleaseDate BETWEEN 1983 AND 1998 OR yearReleaseDate BETWEEN 2006 AND 2010") \
    .select("movieId",upper("title").alias("title"), "yearReleaseDate", "releaseDate", "durationTime")
```

```python
display(movie_day3_df)
```

```python
%sql
CREATE TABLE IF NOT EXISTS movie_demo.movies_merge(
  movieId INT,
  title STRING,
  yearReleaseDate INT,
  releaseDate DATE,
  durationTime INT,
  createdDate DATE,
  updateDate DATE
)
```

####Día 1

```python
%sql
MERGE INTO movie_demo.movies_merge tgt
USING movie_day1 src
ON tgt.movieId = src.movieId
WHEN MATCHED THEN
  UPDATE SET
    tgt.title = src.title,
    tgt.yearReleaseDate = src.yearReleaseDate,
    tgt.releaseDate = src.releaseDate,
    tgt.durationTime = src.durationTime,
    tgt.updateDate = current_timestamp
WHEN NOT MATCHED
  THEN INSERT (tgt.movieId, tgt.title, tgt.yearReleaseDate, tgt.releaseDate, tgt.durationTime, tgt.createdDate)
  VALUES (src.movieId, src.title, src.yearReleaseDate, src.releaseDate, src.durationTime, current_timestamp)
```

```python
%sql
SELECT *
FROM movie_demo.movies_merge;
```

####Día 2

```python
%sql
MERGE INTO movie_demo.movies_merge tgt
USING movie_day2 src
ON tgt.movieId = src.movieId
WHEN MATCHED THEN
  UPDATE SET
    tgt.title = src.title,
    tgt.yearReleaseDate = src.yearReleaseDate,
    tgt.releaseDate = src.releaseDate,
    tgt.durationTime = src.durationTime,
    tgt.updateDate = current_timestamp
WHEN NOT MATCHED
  THEN INSERT (tgt.movieId, tgt.title, tgt.yearReleaseDate, tgt.releaseDate, tgt.durationTime, tgt.createdDate)
  VALUES (src.movieId, src.title, src.yearReleaseDate, src.releaseDate, src.durationTime, current_timestamp)
```

```python
%sql
SELECT *
FROM movie_demo.movies_merge;
```

####Día 3 - con Python

```python
from delta.tables import *

deltaTablePeople = DeltaTable.forName(spark, 'movie_demo.movies_merge')

deltaTablePeople.alias('tgt') \
  .merge(
    movie_day3_df.alias('src'),
    'tgt.movieId = src.movieId'
  ) \
  .whenMatchedUpdate(set =
    {
      "tgt.title": "src.title",
      "tgt.yearReleaseDate": "src.yearReleaseDate",
      "tgt.releaseDate": "src.releaseDate",
      "tgt.durationTime": "src.durationTime",
      "tgt.updateDate": "CURRENT_TIMESTAMP()"
    }
  ) \
  .whenNotMatchedInsert(values =
    {
      "tgt.movieId": "src.movieId",
      "tgt.title": "src.title",
      "tgt.yearReleaseDate": "src.yearReleaseDate",
      "tgt.releaseDate": "src.releaseDate",
      "tgt.durationTime": "src.durationTime",
      "tgt.createdDate": "CURRENT_TIMESTAMP()"
    }
  ) \
  .execute()
```

```python
%sql
SELECT * FROM movie_demo.movies_merge;
```

####Convertir formato "Parquet" a "Delta"

```python
%sql
CREATE TABLE IF NOT EXISTS movie_demo.movies_convert_to_delta(
    movieId INT,
    title STRING,
    yearReleaseDate INT,
    releaseDate DATE,
    durationTime INT,
    createdDate DATE,
    updatedDate DATE
)
USING PARQUET
LOCATION 'abfss://demo@moviehistorybymg.dfs.core.windows.net/movies_convert_to_delta'
```

```python
%sql
INSERT INTO movie_demo.movies_convert_to_delta
SELECT * FROM movie_demo.movies_merge;
```

```python
%sql
CONVERT TO DELTA movie_demo.movies_convert_to_delta;
```
