#### Acceder a azure data lake storage mediante token SAS
1. Registrar la apliación en Azure Entra ID / Service Principal
2. Generar un secreto (Contraseña) para la aplicación
3. Configurar Spark con APP / Client Id, Directory / Tenand Id & Secret
4. Asignar el Role "Storage Blob Data Contributor" al Data Lake

```python
cliend_id = dbutils.secrets.get(scope="movie-history-secret-scope", key="client-d")
tenant_id = dbutils.secrets.get(scope="movie-history-secret-scope", key="tenand-id")
client_secret = dbutils.secrets.get(scope="movie-history-secret-scope", key="client-secret")
```

```python
spark.conf.set("fs.azure.account.auth.type.moviehistorybymg.dfs.core.windows.net", "OAuth")
spark.conf.set("fs.azure.account.oauth.provider.type.moviehistorybymg.dfs.core.windows.net", 
               "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")
spark.conf.set("fs.azure.account.oauth2.client.id.moviehistorybymg.dfs.core.windows.net", cliend_id)
spark.conf.set("fs.azure.account.oauth2.client.secret.moviehistorybymg.dfs.core.windows.net", client_secret)
spark.conf.set("fs.azure.account.oauth2.client.endpoint.moviehistorybymg.dfs.core.windows.net", 
               f"https://login.microsoftonline.com/{tenant_id}/oauth2/token")
```

```python
display(dbutils.fs.ls("abfss://demo@moviehistorybymg.dfs.core.windows.net"))
```
