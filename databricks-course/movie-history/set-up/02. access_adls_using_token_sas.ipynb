#### Acceder a azure data lake storage mediante token SAS
1. Establecer la configuración de spark "SAS Token"
2. Listar archivos del contenedor "demo"
3. Leer dato2 del archivo "movie.csv"

```python
movie_sas_token = dbutils.secrets.get(scope="movie-history-secret-scope", key="movie-sas-token")
```

```python
spark.conf.set("fs.azure.account.auth.type.moviehistorybymg.dfs.core.windows.net", "SAS")
spark.conf.set("fs.azure.sas.token.provider.moviehistorybymg.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.sas.FixedSASTokenProvider")
spark.conf.set("fs.azure.sas.fixed.token.moviehistorybymg.dfs.core.windows.net", movie_sas_token)
```

```python
display(dbutils.fs.ls("abfss://demo@moviehistorybymg.dfs.core.windows.net"))
```
