#### Acceder a azure data lake storage mediante access key
1. Establecer la configuración de spark "fs.azure.account.key"
2. Listar archivos del contenedor "demo"
3. Leer dato2 del archivo "movie.csv"

```python
movie_access_key = dbutils.secrets.get(scope="movie-history-secret-scope", key="movie-access-key")
```

```python
spark.conf.set(
    "fs.azure.account.key.moviehistorybymg.dfs.core.windows.net", movie_access_key)
```

```python
display(dbutils.fs.ls("abfss://demo@moviehistorybymg.dfs.core.windows.net"))
```

```python
display(spark.read.csv ("abfss://demo@moviehistorybymg.dfs.core.windows.net/movie.csv"))
```
