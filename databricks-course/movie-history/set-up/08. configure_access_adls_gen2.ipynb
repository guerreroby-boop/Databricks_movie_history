####Configurar Acceso Azure Data Lake Stirage Gen2
1. Crear objeto "External Location" llamado "external_location_bronze_mov_his"
2. Crear objeto "External Location" llamado "external_location_silver_mov_his"
3. Crear objeto "External Location" llamado "external_location_gold_mov_his"
4. Validar el acceso a los objetos "Externals Locations" creados

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS external_location_bronze_mov_his
    URL 'abfss://bronze@moviehistorybymg.dfs.core.windows.net'
    WITH (STORAGE CREDENTIAL storage_credential_mov_his)
    COMMENT 'External Location For the Bronze Data Lakehouse'
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS external_location_silver_mov_his
    URL 'abfss://silver@moviehistorybymg.dfs.core.windows.net'
    WITH (STORAGE CREDENTIAL storage_credential_mov_his)
    COMMENT 'External Location For the Silver Data Lakehouse'
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS external_location_gold_mov_his
    URL 'abfss://gold@moviehistorybymg.dfs.core.windows.net'
    WITH (STORAGE CREDENTIAL storage_credential_mov_his)
    COMMENT 'External Location For the Gold Data Lakehouse'
```
